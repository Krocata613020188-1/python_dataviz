```python
from matplotlib import pyplot as plt
import matplotlib
import pandas
import numpy as np
```

# scatter plot

```python
irisdata = pandas.read_csv('C:\\Users\\ASUS\\Desktop\\pythonnew\\iris.data',header=None)
irisdata
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5,:]
```

#### สร้างกราได้โดยการกำหนดค่าที่จะเป็น x และ y

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[1:2])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])
```

#### ใช้ตัวย่อในการเลือกสีได้

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c= 'k')
```

```python
print(f'msx = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else :
        colors.append('r')
colors
```

#### ใช้ฟังก์ชั่นเป็นตัวกำหนดสีได้

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = colors)
```

#### ใช้ cmap เพื่อเลือกสีได้

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='YlOrBr')
```

#### ใช้ c = np.array([[0-1,0-1,0-1]])เพื่อปรับระดีบสีเองได้

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = np.array([[1,0.5,0.8]]))
```

### ใช้ marker เพื่อเปลี่ยนรูปร่างของ mark

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='YlOrBr',marker="P")
```

# visualize 2 data
#### การทพกราฟพร้อมกัน2ข้อมูล สามารถนำมาเขียนต่อกันได้เลย

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='Blues',marker="P")
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2], c = versicolor_np[:,3],cmap='Blues',marker="*")
```

# visualize 3 data
#### การทพกราฟพร้อมกัน3ข้อมูล สามารถนำมาเขียนต่อกันได้เลย


```python
virginica_np = irisdata.iloc[100:150,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='Blues',marker="P")
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2], c = versicolor_np[:,3],cmap='Blues',marker="*")
plt.scatter(virginica_np[:,0],virginica_np[:,1],s=30*virginica_np[:,2], c = virginica_np[:,3],cmap='Blues',marker="+")
```
