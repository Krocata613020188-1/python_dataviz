    - ถ้าค่าใน list แรกเป็นเลขคู่ให้ค่าใน dict ที่เป็น output มีkey เป็นค่าจาก list ตัวที่2 และ value จาก list ตัวที่ 1

#### ป.ล.เขียนการให้ index ใน listเป็นเงื่อนไขจำนวนคู่หรือคี่ไม่เป็นครับ

```python

def smartDict(listnumber, listtext):
    if listnumber != evennumber:
        retdict = {}
        for index in range(len(listnumber)):
            retdict[listnumber[index]] = listtext[index]
    else:
        retdict = {}
        for index in range(len(listnumber)):
            retdict[listtext[index]] = listnumber[index]
    return retdict
```

```python
smartDict([1,2,3],['apple','banana','orange'])
```

### 2>จากไฟล์ wuhan_GNome.txt จงเขียนโปรแกรมเพื่ออบคำถามต่อไปนี้
    - กรดอะมิโนใด (A, T, G, C)มีประมาณมากที่สุดและมีเท่าไร
    - จงหา sequence ที่มีขนาด 3 และ 4 ที่เกิดขึ้นซ้ำมากที่สุด

```python
import pandas as pd 
```

```python
DNAseq = pd.read_csv('C:\\Users\\ASUS\\Desktop\\pythonnew\\Wuhan_nCoV_GNome.txt',header = None)
DNAseq.head()
```

```python
def countChar(seq, character):
    charcount = 0
    for char in seq:
        if char == character:
            charcount = charcount + 1
    return charcount
```

```python
ANum = 0
TNum = 0
GNum = 0
CNum = 0
for row in range(DNAseq.shape[0]):
    ANum += countChar(DNAseq.iloc[row][0],'A')
    TNum += countChar(DNAseq.iloc[row][0],'T')
    GNum += countChar(DNAseq.iloc[row][0],'G')
    CNum += countChar(DNAseq.iloc[row][0],'C')
    
print(f'there are {ANum} As, {TNum} Ts, {GNum} Gs, {CNum} Cs,')  
```

### ตอบข้อ2.1 
### กรดอะมิโนที่มีมากที่สุดคือ T และมีจำนวน 9594 ตัว

```python
def populate3charstring(seq):
    str_list = list()
    for i in range(len(seq)-2):
        str_list.append(seq[i:i+3])
    return str_list
   
```

```python
populate3charstring('ATTAAAGG')
```

```python
list_all_3char = list()
for row in range(DNAseq.shape[0]):
    list_all_3char = list_all_3char + populate3charstring(DNAseq.iloc[row][0])
```

```python
len(list_all_3char)
```

```python
unique_3char = list(set(list_all_3char))
len(unique_3char)
```

```python
unique_3char
```

```python
dict_3char = {}
for sub_str in unique_3char:
    num = 0
    for s in list_all_3char:
        if sub_str == s:
            num += 1
    dict_3char[sub_str] = num
```

```python
dict_3char
```

```python
maxVal = -100
maxKey = None
for k in dict_3char.keys():
    if dict_3char[k] > maxVal:
        maxVal = dict_3char[k]
        maxKey = k
```

```python
maxKey
```

```python
print(f'max 3-long-sequence is {maxKey} with frequency equal {maxVal}')
```
